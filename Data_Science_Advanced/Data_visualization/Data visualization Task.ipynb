#### 1. Using a line plot visualize the Total profit data provided for each month.

**Note:**It must be implemented in three methods
- Generated line plot must include the following properties:

X axis name = Month Number <br>
Y axis name = Total profit <br>
The line plot graph should look like this.


```python
#write your code here
# Convert 'total_profit' column to a list
profitList = df['total_profit'].tolist()

# Convert 'month_number' column to a list
monthList = df['month_number'].tolist()

# Plot the profit trends
plt.plot(monthList, profitList, linestyle='-', label='Profit')

# Set x-axis label
plt.xlabel('Month Number')

# Set y-axis label
plt.ylabel('Total profit')

# Set title for the plot
plt.title('Company Profit per Month')

# Set x-axis ticks
plt.xticks(monthList)

# Set y-axis ticks
plt.yticks([100000, 200000, 300000, 400000, 500000])

# Show legend
plt.legend()

# Display the plot
plt.show()
```

#### 2. Read all product sales data and display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).
**Note:**It must be implemented in three methods  
The graph should look like this.


```python
#write your code here
# Drop unnecessary columns
columns = df.columns.drop(['month_number', 'total_units', 'total_profit'])

# Convert 'month_number' column to a list
monthList = df['month_number'].tolist()

# Create subplots
fig, ax1 = plt.subplots()

# Plot data for each column
for column in columns:
    ax1.plot(monthList, df[column], 'o-')

# Set x-axis label
plt.xlabel('Month Number')

# Set y-axis label
plt.ylabel('Sales Units in Number')

# Set title for the plot
plt.title('Sales Data')

# Set x-axis ticks
ax1.set_xticks(monthList)

# Set y-axis ticks
plt.yticks([1000, 2000, 4000, 8000, 10000, 12000, 15000, 18000])

# Set legend labels
legend_labels = ["Face Cream Sales Data", "Face Wash Sales Data", "Toothpaste Sales Data"]

# Display legend
ax1.legend(legend_labels, loc="upper left")

# Show the plot
plt.show()
```

#### 3. Read face cream and facewash product sales data and show it using the bar chart.


- The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.  
- **Note:**It must be implemented in three methods  

The bar chart should look like this.


```python
#write your code here

# Create an array for the x-axis using the length of the "facecream" data
x_axis = np.arange(len(df["facecream"]))

# Set the width of each bar
width = 0.4

# Create a figure and axis object
fig, ax = plt.subplots()

# Plot the bar chart for Face Cream sales data
ax.bar(x_axis - width/2, df["facecream"], width, label='Face Cream sales data')

# Plot the bar chart for Face Wash sales data
ax.bar(x_axis + width/2, df["facewash"], width, label='Face wash sales data')

# Set the x-axis tick positions and labels using the "month_number" data
ax.set_xticks(x_axis)
ax.set_xticklabels(df["month_number"])

# Set the x-axis label, y-axis label, and chart title
ax.set_xlabel("Month Number")
ax.set_ylabel("Sales units in number")
ax.set_title("Facewash and Facecream sales data")

# Add a legend to the upper left corner of the plot
ax.legend(loc="upper left")

# Add gridlines with a linewidth of 1 and dashed linestyle
ax.grid(True, linewidth=1, linestyle="--")

# Display the plot
plt.show()
```

#### 5. Read the total profit of each month and show it using the histogram to see the most common profit ranges.  
**Note:**It must be implemented in three methods.  
The histogram should look like this.


```python
# Define the profit ranges for the histogram bins
PR = [150000, 175000, 200000, 225000, 250000, 300000, 350000]

# Create a figure and axis object
fig, ax = plt.subplots()

# Plot the histogram of total profit using the defined profit ranges
ax.hist(df['total_profit'], bins=PR)

# Add a legend to the upper left corner of the plot
ax.legend(["Profit data"], loc='upper left')

# Set the chart title, x-axis label, and y-axis label
ax.set_title('Profit data')
ax.set_xlabel('Profit range in dollars')
ax.set_ylabel('Actual Profit in dollars')

# Set the x-axis tick positions and the y-axis tick positions
ax.set_xticks([150000, 175000, 200000, 225000, 250000, 300000, 350000])
ax.set_yticks([0, 1, 2, 3, 4, 5])

# Display the plot
plt.show()
```

#### 6. Calculate total sale data for last year for each product and show it using a Pie chart.
- **Note:**
    - In Pie chart display Number of units sold per year for each product in percentage.
    - It must be implemented in two methods.

The Pie chart should look like this.


```python
# Create a figure and axis object with white background
fig, ax = plt.subplots(facecolor='white')

# Define labels and sizes for the pie chart
labels = ["Bathing soap", "Toothpaste", "Facewash", "Facecream", "Moisturizer", "Shampoo"]
sizes = [
    df["bathingsoap"].sum(),
    df["toothpaste"].sum(),
    df["facewash"].sum(),
    df["facecream"].sum(),
    df["moisturizer"].sum(),
    df["shampoo"].sum()
]

# Plot the pie chart with labels, percentage values, shadow, and starting angle of 90 degrees
ax.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)

# Set the chart title
ax.set_title('Sales data')

# Add a legend to the upper right corner of the plot with labels placed outside the pie chart
ax.legend(labels, loc="upper right", bbox_to_anchor=(1, 0, 0.5, 1))

# Display the plot
plt.show()
```
